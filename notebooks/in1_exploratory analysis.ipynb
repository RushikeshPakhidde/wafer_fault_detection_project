```python
def get_cols_with_zero_std_dev(df:pd.DataFrame):
    """Return a list of columns names who are having zero standard deviation"""
    cols_to_drop=[]
    num_cols=[col for col in df.columns if df[col].dtype != 'O']
    #numericals cols only
    for col in num_cols:
        if df[col].std()==0:
            cols_to_drop.append(col)
    return cols_to_drop

def get_redundant_cols(df:pd.DataFrame,missing_thresh=.7):
    """Return a list of columns having missing values more than certain thresh."""
    cols_missing_ratios=df.isna().sum().div(df.shape[0])
    cols_to_drop=list(cols_missing_ratios[cols_missing_ratios>missing_thresh].index)
    return cols_to_drop
```

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
from typing import Tuple
from dataclasses import dataclass
@dataclass
class ClusterDataInstances:
    """Divides tyhe given data instances into different clusters via KMeans Clustering algorithm.
    Args:
        x(np.array):Takes in an array which gotta be clustered.
        desc(str):Description of the said array
        """
    x:np.array
    desc:str

    def _get_ideal_numbers_of_clusters(self):
        """Returns the ideal number of clusters the given data instance should be divided into by locating the dispersa point 
        in number of clusters vs WCSS plot.
        Raise:
             e:Raise relevent exception should any kinda error  pops up while determining the ideal number of clusters.
        Returns:
            int:Ideal numbers of clusters the given data instances should be divided into.
        """
        try:
            print(
                f'Getting the ideal number of clusters to clusters.."{self.desc}set"into..')
            #################### Compute WCSS For shortlisted number of clusters##########
            print('computing Wcss for shortlisteds number of clusters..')
            wcss=[]# within summetion of squares
            for i in range(1,11):
                kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
                kmeans.fit(self.x)
                wcss.append(kmeans.inertia_)
                print(f'WCSS for n_clusters={i}:{kmeans.inertia_}')
            print(
                'WCSS  computed sucessfully for all shortlisted number of clusters!')
                ############# Finalize Dispersal point as the ideal number of clusters############
            print(
                'Finding the ideal number of clusters(by locatinmg thye dispersal point)via Elbow Method')
            knee_finder=KneeLocator(
                range(1,11),wcss,curve='convex',
                direction='decreasing'
                )
            print(
                f'Ideal number of clusters to be formed:{knee_finder.knee}')
            return knee_finder.knee
            ...
        except Exception as e:
            print(e)
            raise e
        
    def create_clusters(self)-> Tuple:
        """Divides the given data instances into the different clusters,the first hand shoud've been divided into
          via offcourse Kmeans clustering algorithm.
          Raises:
          e:Raises relevent exception should anyt kind error pops up while dividing the given data instaces into clusters.
          Returns:
          (KMeans,np.array):KMeans Clustering object being used to cluster the given data instance and the given dataset along 
          with the cluster labels,respectively."""
        try:
            ideal_clusters=self._get_ideal_numbers_of_clusters()
            print(
                f"Dividing the \"{self.desc}\" instance into {ideal_clusters} clusters via KMeans Clustering algorithm..")
            kmeans=KMeans(n_clusters=ideal_clusters,init='k-means++',random_state=42)
            y_kmeans=kmeans.fit_predict(self.x)
            print(
                f"..said data instance divided into{ideal_clusters} clusters sucessfully!")
            return kmeans,np.c_[self.x,y_kmeans]
            ...
        except Exception as e:
            print(e)
            raise e
```

```python
from imblearn.combine import SMOTETomek

x,y=x_trans[:,:-1],y
resampler=SMOTETomek(sampling_strategy='auto')
x_res,y_res=resampler.fit_resample(x,y)
```
